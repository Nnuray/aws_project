```python
# Названия колонок
print(df.columns.tolist())

# Общая информация
df.info()

# Статистика по числовым колонкам
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df["Calories"], bins=20, kde=True)
plt.title("Распределение калорий")
plt.xlabel("Калории")
plt.ylabel("Количество блюд")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Category", y="Calories", data=df, estimator='mean')
plt.title("Среднее количество калорий по категориям")
plt.xticks(rotation=45)
plt.show()
```

```python
# Пример: блюда меньше 500 калорий
low_cal = df[df["Calories"] < 500]
print(low_cal[["Item", "Category", "Calories"]])

# Пример: без сахара
no_sugar = df[df["Sugars"] == 0]
print(no_sugar[["Item", "Category", "Sugars"]])

# Пример: категория десерты (если есть такие)
desserts = df[df["Category"].str.contains("Dessert", case=False)]
print(desserts)
```
